```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df = pickle.load(open('benchmark_results.pckl', 'rb'))
df['i'] = list(range(len(df['gridincrements'].values)))
df = df.set_index(['i'])
df['log_loss'] = df['loss'].apply(np.log)
df = df[['mode', 'neigbors', 'gridincrements', 'timing', 'loss', 'log_loss']]
```
